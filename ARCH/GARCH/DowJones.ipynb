```python
import yfinance as yf
import pandas as pd 
import numpy as np

ticker = yf.Ticker("^DJI")
df = ticker.history(start="2011-03-01", end="2021-03-25", interval="1d")
#df.to_csv("DJI_history.csv")

# filtering the dataframe to only have the date and close price 
df = df[["Close"]]

#adding daily return column
df["Daily_Return"] = df["Close"].pct_change()
```

```python
# filtering the dataframe to only have the date and close price 
df = df[["Close"]]

#adding daily return column and log return column 
df["Daily_Return"] = df["Close"].pct_change()
df["Log_Return"] = np.log(1 + df["Daily_Return"])
```
